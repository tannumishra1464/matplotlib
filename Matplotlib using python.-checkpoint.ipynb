# Basic Charting Using Matplotlib.

```python
%matplotlib in line
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3], [2,5,7])
plt.title('Tannus Data')
plt.ylabel('Y axia Data')
plt.show()
```

```python
plt.plot([1,2,3], [2,5,7])
plt.title('Tannus Data')
plt.xlabel('This is "X" axis')
plt.ylabel('This is "Y" axis')
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use("ggplot")
x=[1,5,4]
y=[23,45,67]
x2=[3,7,3]
y2=[5,8,8]
plt.scatter(x,y, label='first')
plt.scatter(x2,y2, label='second')
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use("ggplot")
x=[1,5,4]
y=[23,45,67]
x2=[3,7,3]
y2=[5,8,8]
plt.plot(x,y, label='first')
plt.plot(x2,y2, label='second')
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use("ggplot")
x=[10,50,60]
y=[23,45,67]
x2=[5,10,15]
y2=[5,8,8]
plt.hist(x,y, label='first')
plt.hist(x2,y2, label='second')
plt.legend()
plt.show()
```
